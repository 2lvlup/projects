```python
import numpy as np
import matplotlib.pyplot as plt

SO = 596.15     # last initial stock price
mu = 0.10        # is the expected price growth
sigma = 0.2     # Volatility (standard deviation)
T = 1           # Time in years
dt = 1/252      # Daily Steps
N = int(T/dt)   # Number of steps
simulations = 5 # Number of sims

# Simulate Paths
for i in range (simulations):
    prices = [SO]
    for _ in range(N):
        shock = np.random.normal(loc=(mu*dt), scale=(sigma*np.sqrt(dt)))
        prices.append(prices[-1] * np.exp(shock))
    plt.plot(prices)

plt.title("Monte Carlo Stock Price Sim")
plt.xlabel("Days")
plt.ylabel("Price")
plt.show()
```
